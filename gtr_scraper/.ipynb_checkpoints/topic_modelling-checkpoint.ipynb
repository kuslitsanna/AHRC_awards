```python
import nltk
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
from gensim.corpora import Dictionary
from gensim.models import ldamodel
from gensim.models.coherencemodel import CoherenceModel
from wordcloud import WordCloud
```

```python
from PIL  import Image
import numpy as np
import random
import re
import matplotlib.pyplot as plt
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('vader_lexicon')
nltk.download('wordnet')
```

```python
text['AllText'] = text['AllText'].str.lower()
text['AllText'] = text['AllText'].str.replace(r'[^A-Za-z\s]','', regex = True)
```

```python
print(text)
```

```python
# Tokenize the corpus (you can add more preprocessing steps here)
abstracts_list = text["AllText"].tolist()
tokenized_abstracts = []

# Tokenize each abstract and append the result to the list
for abstract in abstracts_list:
    tokens = word_tokenize(abstract)
    tokenized_abstracts.append(tokens)
```

```python
from tqdm import tqdm
# Define your list of stopwords
custom_stopwords = set(["project", "research", "also", "within", "including", "well", "across", "ways", "work", "study", "theme", "via", "one", "two", "three", "ahrc", "funder", "researcher", "intelligence"])
stop_words = set(stopwords.words("english"))|custom_stopwords

# Create an empty list to store the filtered tokenized abstracts
filtered_tokenized_abstracts = []

# Iterate through the tokenized abstracts and filter out stopwords with tqdm
for tokens in tqdm(tokenized_abstracts, desc="Filtering Stopwords"):
    filtered_tokens = [word for word in tokens if word not in stop_words]
    filtered_tokenized_abstracts.append(filtered_tokens)
```

```python
filtered_tokenized_abstracts = [
    [word for word in tokens if len(word) >= 3]
    for tokens in filtered_tokenized_abstracts
]
```

```python
# Initialize the WordNet lemmatizer
lemmatizer = WordNetLemmatizer()

# Create an empty list to store the lemmatized tokenized abstracts
lemmatized_tokenized_abstracts = []

# Iterate through the filtered tokenized abstracts and lemmatize with tqdm
for tokens in tqdm(filtered_tokenized_abstracts, desc="Lemmatizing Tokens"):
    lemmatized_tokens = [lemmatizer.lemmatize(word) for word in tokens]
    lemmatized_tokenized_abstracts.append(lemmatized_tokens)
```

```python
print(lemmatized_tokenized_abstracts[:1])
```

```python
%store lemmatized_tokenized_abstracts

import pickle

# Save the list to a file
with open('lemmatized_tokenized_abstracts.pkl', 'wb') as file:
    pickle.dump(lemmatized_tokenized_abstracts, file)

```

```python
# Convert each abstract into a string
abstracts_as_strings = [' '.join(abstract) for abstract in lemmatized_tokenized_abstracts]

# Now, you can join the abstracts as a single string
corpus = ' '.join(abstracts_as_strings)

# Tokenize the corpus
corpus = word_tokenize(corpus)

stop_words = set(stopwords.words("english"))

filtered_corpus = [word for word in corpus if word not in stop_words]

filtered_corpus = list(filtered_corpus)
filtered_corpus = ', '.join(filtered_corpus)


wordcloud = WordCloud(max_words=100, stopwords = stop_words, random_state = 1).generate(filtered_corpus)
plt.figure(figsize=(10,16))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
# Tokenize the filtered_corpus into words
tokens = word_tokenize(filtered_corpus)
tokens = [token for token in tokens if token != ',']

# Create the frequency distribution
fdist = nltk.FreqDist(tokens)

# Get the most common words (top 100)
fdist.most_common(100)
```

```python
plt.figure(figsize=(12,6))
fdist.plot(70)
```

```python
dictionary = Dictionary(lemmatized_tokenized_abstracts)
len(dictionary)
```

```python
# Create a Document-Term Matrix (DTM) using the dictionary
dtm = [dictionary.doc2bow(doc) for doc in lemmatized_tokenized_abstracts]
```

```python
np.random.seed(1)
k_range = range(15,25,2)
scores = []
for k in k_range:
    lda_model = ldamodel.LdaModel(dtm, num_topics=k, id2word=dictionary, passes=20)
    cm = CoherenceModel(model = lda_model, corpus = dtm, dictionary = dictionary, coherence = "u_mass")
    print(cm.get_coherence())
    scores.append(cm.get_coherence())
plt.figure()
plt.plot(k_range, scores)
```

```python
from sklearn.model_selection import train_test_split

train_df, valid_df = train_test_split(dtm, test_size=0.5, random_state=42)


# Specify the number of topics you want to extract
num_topics = 17 # Adjust the number of topics as needed


# Initialize LDA model
lda_model = ldamodel.LdaModel(corpus=train_df, id2word=dictionary, num_topics=num_topics, iterations=500)
```

```python
num_topics = lda_model.num_topics  # Get the total number of topics in the model

# Print all identified topics with their top terms
for topic_id in range(num_topics):
    print(f"Topic {topic_id + 1}:")
    print(lda_model.show_topic(topic_id, topn = 20))
    print()
```

```python
document_topic_distributions = [lda_model.get_document_topics(doc) for doc in dtm]

%store lda_model document_topic_distributions
```

```python
print(document_topic_distributions[:10])
```

```python
print(len(document_topic_distributions))
```

```python
df = pd.DataFrame(index=range(len(document_topic_distributions)), columns=range(17))

# Iterate through the list of tuples and populate the DataFrame
for row_idx, item in enumerate(document_topic_distributions):
    for x, y in item:
        df.at[row_idx, x] = y
        
df = pd.DataFrame(df, dtype=float)

# Round the values to two decimal places
df = df.round(2)

df[df < 0.1] = np.nan

# Convert column names from strings to integers
df.columns = df.columns.astype(int)

# Create a mapping from old column names (integers) to new column names
column_mapping = {i: f'Topic {i + 1}' for i in range(32)}

# Rename the columns using the mapping
df = df.rename(columns=column_mapping)

print(df[:5])

```
